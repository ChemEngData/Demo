```python
import matplotlib.pyplot as plt
from skimage import io, color
from skimage.feature import canny
from skimage.filters import sobel
```

This ios some text

```python
image_path = 'input/turtle.jpg'
image = io.imread(image_path)

gray_image = color.rgb2gray(image)
edges = canny(gray_image, sigma=2)  # Adjust sigma as needed for your image

plt.figure(figsize=(10, 5))

# Original image
plt.subplot(1, 2, 1)
plt.imshow(image)
plt.title('Original Image')
plt.axis('off')

# Edges
plt.subplot(1, 2, 2)
plt.imshow(edges, cmap='gray')
plt.title('Edges')
plt.axis('off')

plt.tight_layout()
plt.show()
```
